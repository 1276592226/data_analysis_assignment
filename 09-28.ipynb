# Learn the data analsysi
0. 数据读取
   ```
   df = pd.read_excel(path,sheet_name=<sheetname>)
   df = pd.read_csv(path)
   df = pd.read_json(<parameters>)
   ```
1. 数据清洗， 去除无效数据
    主要算法：
    ```
    def condition(value):
        return isinstance(value, (int, float))
    df2 = df2[df2["Profit"].map(condition)]
    ```
    有些字段需要数值型，对于不符合要求的，需要删除
2. 筛选条件
   筛选行
   ```
   df[df[<field>]=='A']
   ```
   筛选列
   ```
   df = df[[<col1>,<col2>,<col3>]]
   ```
3. 数据汇总
    语法
    ```
    df = pd.pivot_table()
    ```
    也可以用教程上的语法
    ```
    df = df.group(<col>).sum()
    ```
4. 相似度/相关系数计算，哪些参数有因果关系
    df有corr方法，可以计算相关系数
    ```
    corr = df.corr()
    ```
5. 推荐算法
    ```
    from apyori import apriori
    results = apriori(order_item_list, min_support=0.005,               min_confidence=0.25)
    print(results)
    ```
